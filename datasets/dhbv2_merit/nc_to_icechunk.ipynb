```python
import icechunk as ic
import xarray as xr
from icechunk.xarray import to_icechunk
```

```python
ds = xr.open_dataset("./merit_dhbv2_retrospective.nc")
```

```python
local_path = "./merit_dhbv2_retrospective"
storage_config = ic.local_filesystem_storage(local_path)
repo = ic.Repository.create(storage_config)
session = repo.writable_session("main")

to_icechunk(ds, session)

snapshot = session.commit("initial MERIT streamflow commit - all zones combined")
print(f"Data saved to {local_path}")
print(f"Commit: {snapshot}")
```
